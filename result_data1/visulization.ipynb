```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import palettable
from pandas import Series,DataFrame
from sklearn import datasets
import seaborn as sns
```

```python
df = pd.read_csv("./result.csv")
df
```

```python
# df1=df
# df.rename(columns={"budgets":"budgets(%)"})
cora=df.loc[df["dataSets"] == 'Cora']
citesee=df.loc[df["dataSets"] == 'CiteSeer']
cora["attacks"]
```

```python
attacks = ["CLGA","dice","metattack","minmax","nodeembeddingattack","pgd","random"]
budgets=["01","05","10","15","20"]
metrics=["edge_homophily","node_homophily","class_homophily","adjusted_homophily","balanced_homophily"]

for me in metrics:
    cora_temp = cora.loc[df["metrics"] == me]
    title="cora_{}".format(me)

    fig=sns.relplot(x='budgets(%)',y='result',data=cora_temp,
                palette=palettable.cartocolors.qualitative.Bold_9.mpl_colors[:7],
                height=7,
                hue='attacks',#æ·»åŠ åˆ†ç±»å˜é‡
                style='attacks',#æ·»åŠ ä¸åŒç±»å˜é‡æŒ‰ç…§ä¸åŒmarkeræ˜¾ç¤º
    #             markers=['d','*','X',".",""],#è‡ªå®šä¹‰markerå½¢çŠ¶ï¼Œå‚è€ƒ
                **dict(s=120),#ä¼ å…¥è¯å…¸ï¼Œè°ƒç”¨Matplotlibä¸­çš„å‚æ•°            
               )
    plt.title(title)
    sns.set(style="darkgrid",font_scale=1.5)
    # scatter_fig = fig.get_figure()
    fig.savefig("./{}.png".format(title), dpi = 400)
```
